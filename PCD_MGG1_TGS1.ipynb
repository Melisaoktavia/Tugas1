```python
# ==========================================
# EKSPLORASI DIGITALISASI CITRA
# ==========================================

# Install OpenCV jika belum ada
!pip install opencv-python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ==========================================
# 1. UPLOAD CITRA
# ==========================================
print("Silakan upload citra hasil digitalisasi...")
uploaded = files.upload()

image_path = list(uploaded.keys())[0]
img = cv2.imread(image_path)

if img is None:
    print("Gambar gagal dibaca!")
    raise Exception("Proses dihentikan karena gambar tidak terbaca.")

print("Gambar berhasil dibaca!")

# Konversi BGR ke RGB untuk tampilan
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Citra Digital")
plt.axis("off")
plt.show()

# ==========================================
# 2. REPRESENTASI MATRIKS & VEKTOR
# ==========================================

print("\n===== REPRESENTASI MATRIKS (5x5 Piksel Pertama) =====")
matrix_5x5 = img[0:5, 0:5]
print(matrix_5x5)

print("\n===== REPRESENTASI VEKTOR (50 Elemen Pertama) =====")
vector = img.flatten()
print(vector[:50])

print("\nPenjelasan:")
print("- Matriks mempertahankan struktur 2D (baris dan kolom piksel).")
print("- Vektor adalah representasi 1D hasil flatten seluruh piksel.")

# ==========================================
# 3. ANALISIS PARAMETER CITRA
# ==========================================

height, width, channels = img.shape
bit_depth = img.dtype.itemsize * 8
aspect_ratio = width / height

print("\n===== ANALISIS PARAMETER CITRA =====")
print(f"Resolusi Spasial : {width} x {height} piksel")
print(f"Jumlah Channel   : {channels}")
print(f"Bit Depth        : {bit_depth} bit per channel")
print(f"Total Bit/Pixel  : {bit_depth * channels} bit")
print(f"Aspect Ratio     : {aspect_ratio:.2f}")

# Jumlah intensitas
intensitas = 2 ** bit_depth
print(f"Jumlah Intensitas per Channel : {intensitas}")

# Ukuran memori
memory_bytes = img.nbytes
memory_kb = memory_bytes / 1024
memory_mb = memory_kb / 1024

print(f"Ukuran Memori : {memory_kb:.2f} KB ({memory_mb:.2f} MB)")

# ==========================================
# 4. SIMULASI PERUBAHAN RESOLUSI & BIT DEPTH
# ==========================================

new_width = width * 2
new_height = height * 2
new_bit_depth = bit_depth / 2

new_memory = (new_width * new_height * channels * new_bit_depth) / 8
new_memory_kb = new_memory / 1024
new_memory_mb = new_memory_kb / 1024

print("\n===== SIMULASI PERUBAHAN =====")
print(f"Resolusi Baru : {new_width} x {new_height}")
print(f"Bit Depth Baru : {new_bit_depth} bit")
print(f"Perkiraan Ukuran Memori Baru : {new_memory_kb:.2f} KB ({new_memory_mb:.2f} MB)")

# ==========================================
# 5. MANIPULASI DASAR (SUDAH DIPERBAIKI)
# ==========================================

# Cropping (ambil bagian tengah gambar)
start_y = int(height * 0.25)
end_y   = int(height * 0.75)
start_x = int(width * 0.25)
end_x   = int(width * 0.75)

crop = img[start_y:end_y, start_x:end_x]

# Resizing (50%)
resize = cv2.resize(img, (width//2, height//2))

# Rotasi 90 derajat
rotate = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

# Flip Horizontal
flip = cv2.flip(img, 1)

# ==========================================
# 6. TAMPILKAN HASIL MANIPULASI
# ==========================================

def show_image(title, image):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.title(title)
    plt.axis("off")
    plt.show()

print("\n===== HASIL MANIPULASI =====")

show_image("Original", img)
show_image("Cropping (Tengah)", crop)
show_image("Resizing 50%", resize)
show_image("Rotasi 90 Derajat", rotate)
show_image("Flip Horizontal", flip)

print("\nEksperimen manipulasi selesai tanpa error.")
```
